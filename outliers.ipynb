```python
import pandas as pd
```

```python
dd = pd.read_csv("datasets/bad_daraz_dataset (1).csv")
dd
```

```python
dd["price"] = pd.to_numeric(dd["price"],errors="coerce")

#Quartiles(Q3), Q1 --> IQR
```

```python
mean = dd["price"].mean()
std = dd["price"].std()
# mean 13007.692307692309
# std #27605.71986438262
dd["z_score"] = (dd["price"]-mean)/std #zscore > 2 --> outliers, <2 --> usual
outliers = dd[dd["z_score"].abs() >  2]
outliers
```

```python
Q1 = dd["price"].quantile(0.25)
Q3 = dd["price"].quantile(0.75)
Median = dd["price"].median()

IQR = Q3-Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

```

```python
outliers = dd[(dd["price"] < lower_bound)|(dd["price"]>upper_bound)]
outliers
```

```python
len(outliers)
```

```python
dd.loc[(dd["price"] < lower_bound) | (dd["price"]> upper_bound),"price"] = Median
dd
```

```python
dd["price"] = dd["price"].clip(lower_bound,upper_bound)
dd
```
