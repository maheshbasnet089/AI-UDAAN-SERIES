```python
dd = pd.read_csv("datasets/daraz_dataset.csv")
dd
```

```python
category = dd.groupby("Category")["Product"].count()
plt.figure(figsize=(18,6))
plt.bar(category.index,category.values,color="green",edgecolor="blue")
plt.xlabel("Categories",fontsize=20)
plt.ylabel("Products",fontsize=20)
plt.title("Products & Category Bar Graph")
plt.grid(axis="y",linestyle="--",alpha=1)
```
