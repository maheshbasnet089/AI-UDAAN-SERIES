```python
dd = pd.read_csv("datasets/daraz_dataset.csv")
dd
```

```python
category = dd.groupby("Category")["Product"].count()
plt.figure(figsize=(18,6))
plt.bar(category.index,category.values,color="green",edgecolor="blue")
plt.xlabel("Categories",fontsize=20)
plt.ylabel("Products",fontsize=20)
plt.title("Products & Category Bar Graph")
plt.grid(axis="y",linestyle="--",alpha=1)
```

```python
category = dd.groupby("Category")["Product"].count()
category
```

```python
import matplotlib.pyplot as plt 
colors = [
    "#000", 
    "brown", 
    "blue", 
    "red", 
    "orange", 
    "purple",
    "pink"
]
plt.figure(figsize=(14,6))
plt.pie(category.values,labels=category.index,colors=colors,shadow=True, startangle=90,explode=(0.1,0,0,0,0,0,0.1))
plt.title("Product Distribution by category")
```

```python
gender_products = dd.groupby("Gender")["Product"].count()
plt.pie(gender_products.values,labels=gender_products.index,colors=["pink","green"])
plt.title("Product By Gender")
```

```python
plt.hist(dd["UnitPrice"],color="green",edgecolor="black",alpha=0.5)
plt.title("Distribution of UnitPrice")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
```
