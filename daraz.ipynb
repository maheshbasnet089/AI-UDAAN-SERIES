```python
import pandas as pd
```

```python
sd = pd.DataFrame({
    "Salary" : [1000,2000,3000,5000,7000,8000,9000,12000,50000]
})

#quarties 
median = sd["Salary"].quantile(0.50)
q1 = sd["Salary"].quantile(0.25)
q3 = sd["Salary"].quantile(0.75)

q3
#IQR 
iqr = q3-q1

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

upper_bound
```

```python
outliers = sd[(sd["Salary"] < lower_bound)|(sd["Salary"] > upper_bound)]
print(outliers)
```
